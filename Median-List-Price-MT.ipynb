```python
df = pd.read_csv('Metro_mlp_uc_sfr_sm_month.csv')
```

```python
selected_columns = ['StateName'] + [col for col in df.columns if any(year in col for year in ['2018', '2019', '2020', '2021', '2022', '2023'])]
selected_data = df[selected_columns]
selected_data
```

```python
# Filter the data for proper states
# stuff=[0,50000,100000,150000,200000,250000,300000,350000,400000,450000]
selected_data_ny_tx = selected_data[selected_data['StateName'].isin(['CA', 'FL', 'NY', 'PA', 'TX'])]
# Create a new column for each year and add the values for each
for year in range(2018, 2024):
    selected_data_ny_tx[f'Total_{year}'] = selected_data_ny_tx.apply(lambda row: row.filter(like=str(year)).sum(), axis=1)
    
# Group by state and select the total for each year
grouped_df = selected_data_ny_tx.groupby('StateName')[[f'Total_{year}' for year in range(2018, 2024)]].sum()

# Plot the bar chart
ax = grouped_df.plot(kind='bar', figsize=(14, 8))
ax.set_ylabel('List Price')
ax.set_xlabel('State')
ax.set_title('List Price by State (2018 to 2023)')
# ax.set_ylabel(stuff)
# Save the chart as an image (PNG format in this example)
plt.savefig('bar_chart.png')
plt.show()
```

```python
#filter data to year 
# Define the range for the year 2018
start_date = pd.Timestamp('2018-01-01')
end_date = pd.Timestamp('2018-12-31')

# Use list comprehension to filter columns within the range of the year 2018
filtered_columns = list(df_pennsylvania.columns[:5]) + [column for column in first_read.columns if start_date <= pd.to_datetime(column, errors='coerce') <= end_date]

# Create a new DataFrame using the filtered columns
df_2018 = df_pennsylvania[filtered_columns]

df_2018['avg_of_2018'] = (df_2018['2018-03-31'] + df_2018['2018-04-30'] + df_2018['2018-05-31'] + df_2018['2018-06-30'] + df_2018['2018-07-31'] + df_2018['2018-04-30'] + df_2018['2018-08-31'] + df_2018['2018-09-30'] + df_2018['2018-10-31'] + df_2018['2018-11-30'] + df_2018['2018-12-31']) / 11

# Display the new DataFrame
df_2018.head()
```

```python
# Specify the specific years and StateNames you are interested in
selected_years = [2018, 2019, 2020, 2021, 2022, 2023]  
selected_states = ['CA', 'NY', 'TX', 'FL', 'PA']  

# Filter DataFrame for selected years and StateNames
filtered_df = first_read[(first_read['Year'].dt.year.isin(selected_years)) & (df['StateName'].isin(selected_states))]

# Plot the results
plt.figure(figsize=(12, 6))

for state in selected_states:
    state_data = filtered_df[filtered_df['StateName'] == state]
    plt.plot(state_data['Year'], state_data['ListPrice'], label=state, marker='o')

plt.title('List Prices for Specific StateNames and Years')
plt.xlabel('Year')
plt.ylabel('List Price')
plt.legend(title='StateName', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

```python
# Filter by the array of states and years
selected_states = ['California', 'New York', 'Texas', 'Florida', 'Pennsylvania']
selected_years = [2018, 2019, 2020, 2021, 2022, 2023]

filtered_df = df[df['StateName'].isin(selected_states)]
filtered_df = filtered_df[filtered_df.columns[filtered_df.columns.year.isin(selected_years)]]

# Calculate the average median list price for each state
average_prices = filtered_df.groupby('StateName').median().mean(axis=1) / 1000  # in thousands

# Plot the bar chart
plt.bar(average_prices.index, average_prices)
plt.title('Average Median List Price by State')
plt.xlabel('StateName')
plt.ylabel('Average Median List Price (in thousands)')
plt.show()
```
